# 📌 Median

## Definition

The **Median** is the middle value of a dataset after arranging the data in ascending (or descending) order.

Unlike the mean, the median is **not affected much by outliers**, making it a better measure of central tendency for skewed data.

---

## Steps to Calculate Median

### Case 1: Odd Number of Observations

1. Arrange the data in ascending order.
2. Find the middle value.

### Example

Dataset:

```
9, 2, 5, 7, 1
```

After Sorting:

```
1, 2, 5, 7, 9
```

The middle value is **5**.

**Median = 5**

---

### Case 2: Even Number of Observations

If the dataset has an even number of observations, there is no single middle value.

Take the average of the two middle values.

Example:

Dataset:

```
2, 4, 6, 8
```

Middle Values:

```
4 and 6
```

Median:

```
(4 + 6) / 2 = 5
```

**Median = 5**

---

## Why is Median Important?

Median is useful when the dataset contains **outliers**.

For example, salaries:

```
24,000
25,000
26,000
27,000
500,000
```

Median:

```
26,000
```

Although one salary is extremely high, the median still represents the typical salary.

This is why data scientists often prefer the **median** over the **mean** when analyzing income, house prices, or other skewed datasets.

````python
import numpy as np

data = [9, 2, 5, 7, 1]

print("Original Dataset:", data)
````

````python
sorted_data = sorted(data)

print("Sorted Dataset:", sorted_data)
````

````python
median = np.median(data)

print("Median =", median)
````

## Observation

The dataset must always be **sorted** before identifying the median.

For an odd number of observations, the median is simply the middle value.

````python
data_even = [2, 4, 6, 8]

print("Dataset:", data_even)
print("Median =", np.median(data_even))
````

## Observation

Since the dataset contains an even number of values, there are two middle elements.

Middle Values:

```
4 and 6
```

Median:

```
(4 + 6) / 2 = 5
```

````python
salary = [24000, 25000, 26000, 27000, 500000]

print("Salaries:", salary)
print("Mean:", np.mean(salary))
print("Median:", np.median(salary))
````

## 📊 Mean vs Median

Output:

- Mean = 120400
- Median = 26000

### Observation

The **Mean** is heavily influenced by the salary of ₹500000.

The **Median** remains close to the salaries earned by most employees.

This is why the median is preferred for analyzing:

- Salaries
- House Prices
- Income Distribution
- Skewed Data

# ✅ Key Takeaways

- Median is the middle value of a sorted dataset.
- Always sort the data before finding the median.
- If the number of observations is odd, choose the middle value.
- If the number of observations is even, take the average of the two middle values.
- Median is resistant to outliers.
- Median is widely used in Data Science for skewed datasets.
