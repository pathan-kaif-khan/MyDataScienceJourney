# 1️⃣ Mean

## Definition

The **Mean** is the average of all observations.

### Formula

\[
Mean=\frac{\sum X}{N}
\]

where

- ΣX = Sum of all observations
- N = Number of observations

### Example

Dataset:

```
2,4,6,8,10
```

Step 1

```
Sum = 30
```

Step 2

```
30 / 5 = 6
```

````python
mean = np.mean(data)

print("Mean =", mean)
````

### Observation

The average value of the dataset is **6**.
