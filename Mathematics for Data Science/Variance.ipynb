# 📌 Variance

## Definition

Variance is a statistical measure that tells us how much the values in a dataset differ from the mean.

In simple words:

> Variance measures the spread or dispersion of data.

---

## Why do we need Variance?

Consider two datasets:

Dataset A:

```
10, 10, 10, 10, 10
```

Dataset B:

```
2, 6, 10, 14, 18
```

Both have:

```
Mean = 10
```

But:

- Dataset A values are close to the mean.
- Dataset B values are far away from the mean.

Variance helps us measure this difference.

---

## Formula

For population variance:

\[
Variance = \frac{\sum (X-\mu)^2}{N}
\]

Where:

- X = Individual value
- μ = Mean of dataset
- N = Number of observations

---

## Steps to Calculate Variance

1. Find the mean.
2. Calculate deviation:

```
Deviation = Value - Mean
```

3. Square each deviation.
4. Find the average of squared deviations.

That final value is the variance.

````python
import numpy as np
import pandas as pd
````

````python
data = [2,4,6,8,10]

print("Dataset:", data)
````

## Step 1: Calculate Mean

Dataset:

```
2,4,6,8,10
```

Mean:

\[
\frac{2+4+6+8+10}{5}=6
\]

Mean = 6

````python
mean = np.mean(data)

print("Mean =", mean)
````

## Step 2: Calculate Deviations

Formula:

\[
Deviation = X - Mean
\]

| X | Deviation |
|---|---|
|2|-4|
|4|-2|
|6|0|
|8|2|
|10|4|


````python
deviations = []

for x in data:
    deviations.append(x - mean)

print(deviations)
````

## Step 3: Square the Deviations

We square because negative and positive deviations cancel each other.

Example:

```
-4 + (-2) + 0 + 2 + 4 = 0
```

After squaring:

```
16, 4, 0, 4, 16
```

Now all values are positive.

````python
squared_deviations = []

for d in deviations:
    squared_deviations.append(d**2)

print(squared_deviations)
````

## Step 4: Calculate Variance

Formula:

\[
Variance=\frac{Sum\ of\ squared\ deviations}{Number\ of\ values}
\]


Calculation:

```
(16+4+0+4+16)/5

=40/5

=8
```

Variance = 8

````python
variance = sum(squared_deviations) / len(data)

print("Variance =", variance)
````

````python
variance = np.var(data)

print("Variance =", variance)
````

# Shortcut Formula

Another way to calculate variance:

\[
Var(X)=E(X^2)-[E(X)]^2
\]


Meaning:

```
Variance =
Average of squared values
-
Square of average value
```

This formula is commonly used in probability and machine learning mathematics.

````python
x_squared_mean = np.mean(np.array(data)**2)

mean_squared = np.mean(data)**2

variance = x_squared_mean - mean_squared

print("E(X²) =", x_squared_mean)
print("[E(X)]² =", mean_squared)
print("Variance =", variance)
````

# 📊 Comparing Spread

Dataset 1:

```
10,10,10,10,10
```

Dataset 2:

```
2,6,10,14,18
```

Both have:

```
Mean = 10
```

But:

Dataset 1:

```
Variance = 0
```

No spread.

Dataset 2:

Large variance.

Values are spread out.

````python
data1 = [10,10,10,10,10]

data2 = [2,6,10,14,18]


print("Variance of Dataset 1:", np.var(data1))

print("Variance of Dataset 2:", np.var(data2))
````

# 📊 Variance in Data Science

Variance is used in:

## 1. Feature Selection

Features with very low variance contain little information.

Example:

```
Age = 25,25,25,25
```

This feature does not help a model.

---

## 2. Understanding Data Distribution

High variance:

```
Values are spread out
```

Low variance:

```
Values are close together
```

---

## 3. Machine Learning Algorithms

Variance is important in:

- Linear Regression
- Principal Component Analysis (PCA)
- Bias-Variance Tradeoff
- Data Standardization
