# 📌 Standard Deviation

## Definition

Standard Deviation is a measure that tells us how much the values in a dataset are spread out from the mean.

In simple words:

> Standard deviation tells us the average distance of data points from the mean.

---

## Relationship with Variance

Standard Deviation is the square root of variance.

Formula:

\[
Standard\ Deviation = \sqrt{Variance}
\]

or

\[
\sigma = \sqrt{\frac{\sum(X-\mu)^2}{N}}
\]

Where:

- X = Individual data value
- μ = Mean
- N = Number of observations
- σ = Standard deviation

---

## Why do we need Standard Deviation?

Variance is measured in squared units.

Example:

If height is measured in:

```
cm
```

Variance will be:

```
cm²
```

This is difficult to interpret.

Taking the square root gives:

```
cm
```

which is easier to understand.

````python
import numpy as np
import pandas as pd
````

````python
data = [2,4,6,8,10]

print("Dataset:", data)
````

# Steps to Calculate Standard Deviation

Example:

```
2,4,6,8,10
```

### Step 1: Find Mean

Mean:

```
(2+4+6+8+10)/5

=6
```

---

### Step 2: Find Deviations

```
Value - Mean
```

|Value|Deviation|
|-|-|
|2|-4|
|4|-2|
|6|0|
|8|2|
|10|4|

---

### Step 3: Square Deviations

```
16,4,0,4,16
```

---

### Step 4: Find Variance

```
(16+4+0+4+16)/5

=8
```

---

### Step 5: Take Square Root

```
√8 = 2.83
```

Therefore:

```
Standard Deviation = 2.83
```

````python
variance = np.var(data)

print("Variance =", variance)
````

````python
standard_deviation = np.sqrt(variance)

print("Standard Deviation =", standard_deviation)
````

````python
std = np.std(data)

print("Standard Deviation =", std)
````

# How to Interpret Standard Deviation?

Suppose:

Mean = 50

Standard Deviation = 5


This means:

Most values are approximately around:

```
50 ± 5
```

Range:

```
45 to 55
```

A small standard deviation means:

- Values are close to mean
- Data is consistent

A large standard deviation means:

- Values are spread out
- Data has more variation

````python
dataset1 = [10,10,10,10,10]

dataset2 = [2,6,10,14,18]


print("Dataset 1 STD:", np.std(dataset1))

print("Dataset 2 STD:", np.std(dataset2))
````

## Observation

Dataset 1:

```
10,10,10,10,10
```

All values are identical.

Standard deviation:

```
0
```

No variation.


Dataset 2:

```
2,6,10,14,18
```

Values are far from the mean.

Standard deviation is higher.

More spread = Higher standard deviation.

````python
student_A = [80,80,80,80,80]

student_B = [50,70,80,90,100]


print("Student A STD:", np.std(student_A))

print("Student B STD:", np.std(student_B))
````

# Student Performance Example

Both students have:

Mean score = 80


Student A:

```
80,80,80,80,80
```

Very consistent.


Student B:

```
50,70,80,90,100
```

Performance varies a lot.


Standard deviation helps measure consistency.

# 📊 Standard Deviation in Data Science

Standard deviation is used in:

## 1. Data Scaling

Machine learning algorithms perform better when features are on similar scales.

Example:

```
Age: 20-60

Salary: 20,000-500,000
```

Standardization uses:

\[
z=\frac{x-\mu}{\sigma}
\]


---

## 2. Outlier Detection

Values far away from the mean are often detected using standard deviation.

Example:

```
Mean = 50

STD = 5
```

A value:

```
90
```

is unusual because it is many standard deviations away.

---

## 3. Probability Distributions

Standard deviation is important in:

- Normal Distribution
- Gaussian Models
- Machine Learning

# ✅ Key Takeaways

- Standard deviation measures spread around the mean.
- It is the square root of variance.
- It has the same units as the original data.
- Low standard deviation = values are close together.
- High standard deviation = values are spread out.
- It is one of the most important concepts in Data Science.
