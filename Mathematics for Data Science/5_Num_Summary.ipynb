```python
### Removing a outlier ###
import numpy as np
arr = np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
```

```python
Q1 = np.percentile(arr,25)
Q3 = np.percentile(arr,75)
```

```python
Q1
```

```python
Q3 
```

```python
IQR = Q3 - Q1
IQR
```

```python
UF = Q3 + (1.5*IQR)
UF
```

```python
LF = Q1 - (1.5*IQR)
LF
```

```python
for i in arr:
    print(i)
```

```python
l = []
for i in arr:
    if i>= LF and i<=UF:
        l.append(i)
l
```

```python
arr2 = np.array(l)
arr
```

```python
arr2
```

```python
import matplotlib.pyplot as plt
plt.boxplot(arr, vert=False)
plt.show()
```
