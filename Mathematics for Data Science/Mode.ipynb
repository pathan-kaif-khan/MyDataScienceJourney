# 📌 Mode

## Definition

The **Mode** is the value that appears **most frequently** in a dataset.

In simple words:

> Mode tells us the most common or most repeated value.

---

## Example

Dataset:

```
2, 3, 4, 4, 5, 6
```

Frequency of values:

| Value | Count |
|------|------|
| 2 | 1 |
| 3 | 1 |
| 4 | 2 |
| 5 | 1 |
| 6 | 1 |

The value **4** appears the most number of times.

Therefore:

```
Mode = 4
```

---

## Why is Mode Important?

Mode is useful when we want to find the most common category or value.

Examples:

- Most purchased product
- Most common shoe size
- Most common color
- Most popular payment method

Unlike mean and median, mode can be used with **categorical data**.

````python
import numpy as np
import pandas as pd
from statistics import mode
````

````python
data = [2, 3, 4, 4, 5, 6]

print("Dataset:", data)
````

## Finding Frequency

Before finding the mode, we count how many times each value appears.

The value with the highest frequency becomes the mode.

````python
frequency = {}

for value in data:
    if value in frequency:
        frequency[value] += 1
    else:
        frequency[value] = 1

frequency
````

````python
result = mode(data)

print("Mode =", result)
````

## Observation

The number **4** appears twice, while all other values appear only once.

Therefore:

```
Mode = 4
```

````python
data_series = pd.Series([2,3,4,4,5,6])

data_series.mode()
````

````python
# Multiple Modes

Sometimes more than one value can have the highest frequency.

Example:

```
1, 2, 2, 3, 3, 4
```

Frequency:

| Value | Count |
|------|------|
|1|1|
|2|2|
|3|2|
|4|1|

Both 2 and 3 appear twice.

Therefore:

```
Mode = 2 and 3
```

This type of dataset is called **Bimodal**.
````
